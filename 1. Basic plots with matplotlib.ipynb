## 1. Line Plot
With matplotlib, you can create a bunch of different plots in Python. The most basic plot is the line plot. A general recipe is given here.

     import matplotlib.pyplot as plt
     plt.plot(x,y)
     plt.show()

```python
import matplotlib.pyplot as plt
```

```python
year = [1500, 1600, 1700,1750,1800,1850,1900,1950,1999,2008,2010,2012,2050,2150]
pop = [585,660, 710, 791, 978, 1262,1650,2521,6008,6707,6896,7052,9725,9746]
```

```python
plt.plot(year, pop)
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("Year vs Population")
plt.yticks([2000, 4000, 6000, 8000, 10000], ["2B","4B","6B","8B","10B"])
plt.grid(True)
plt.show()
```

```python
# Print the last item from year and pop
print(year[3])
print(pop[3])
```

## 2. Scatter Plot

When you have a time scale along the horizontal axis, the line plot is your friend. But in many other cases, when you're trying to assess if there's a correlation between two variables, for example, the scatter plot is the better choice. Below is an example of how to build a scatter plot.

    import matplotlib.pyplot as plt
    plt.scatter(x,y)
    plt.show()

```python
import matplotlib.pyplot as plt
```

```python
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]
```

```python
plt.scatter(year, pop)
plt.show()
```

## 3. Histograms

    ● Explore dataset
    ● Get idea about distribution

```python
import matplotlib.pyplot as plt
```

```python
values = [0,0.6,1.4,1.6,2.2,2.5,2.6,3.2,3.5,3.9,4.2,6]
```

```python
plt.hist(values) #default bins equals 10
plt.show()
```

By default, **Python sets the number of bins to 10**. 

The number of `bins` is pretty important. 

    ● Too few bins will oversimplify reality and won't show you the details. 
    ● Too many bins will overcomplicate reality and won't show the bigger picture.

```python
plt.hist(values, bins=5)
plt.show()

plt.clf()              #plt.clf() cleans it up again so you can start afresh

plt.hist(values, bins=20)
plt.show()
```

## Labels 

To add x-axis and y-axis labels use the `xlabel()` and  `ylabel()` functions respectively. 

To add title use `title()` function. 

The functions are available in `matplotlib.pyplot`. 

```python
import matplotlib.pyplot as plt
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]


plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')

plt.scatter(year, pop)
plt.plot(year, pop)
plt.show()
```

```python
plt.scatter(year, pop)
plt.clf() 
plt.plot(year, pop)
plt.show()
```

```python
import matplotlib.pyplot as plt
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]


plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')

plt.scatter(year, pop)
plt.show()

plt.clf()    #plt.clf() cleans it up again so you can start afresh

plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
plt.plot(year, pop)
plt.show()
```

## Ticks

Segments on the axises (x-axis or y-axis).

    e.g. plt.yticks([0,1,2], ["one","two","three"])

```python
import matplotlib.pyplot as plt
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]

plt.plot(year, pop)
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')

#ticks

plt.yticks([3,4,5,6,7], ["3 Billion","4 Billion","5 Billion","6 Billion","7 Billion"])
plt.grid(True)
plt.show()
```
